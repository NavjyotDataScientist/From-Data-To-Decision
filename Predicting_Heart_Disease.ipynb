```python
import pandas as pd
import numpy as np

train_path = "/content/train.csv"
test_path  = "/content/test.csv"

train = pd.read_csv(train_path)
test  = pd.read_csv(test_path)

print("Train shape:", train.shape)
print("Test shape :", test.shape)

display(train.head(3))
print("\nColumns:\n", train.columns)

print("\nTrain info:")
display(train.info())

print("\nTarget distribution:")
print(train["Heart Disease"].value_counts())
print("\nTarget %:")
print(train["Heart Disease"].value_counts(normalize=True) * 100)

print("\nMissing values in train:")
display(train.isna().sum().sort_values(ascending=False).head(15))

print("\nMissing values in test:")
display(test.isna().sum().sort_values(ascending=False).head(15))

# Make copies (always safe)
train_clean = train.copy()
test_clean  = test.copy()

# Save ids
train_ids = train_clean["id"]
test_ids  = test_clean["id"]

# Drop id
train_clean.drop(columns=["id"], inplace=True)
test_clean.drop(columns=["id"], inplace=True)

# Target mapping
target_map = {"Absence": 0, "Presence": 1}
train_clean["Heart Disease"] = train_clean["Heart Disease"].map(target_map)

# Check mapping success
print("Target after mapping:")
print(train_clean["Heart Disease"].value_counts(dropna=False))

# Quick sanity checks
print("\nDuplicates in train:", train_clean.duplicated().sum())
print("Duplicates in test :", test_clean.duplicated().sum())

import matplotlib.pyplot as plt

# Numeric summary
display(train_clean.describe().T)

# Correlation with target
corr = train_clean.corr(numeric_only=True)["Heart Disease"].sort_values(ascending=False)
print("Correlation with target:")
display(corr)

# Target distribution plot
train_clean["Heart Disease"].value_counts().plot(kind="bar")
plt.title("Target Distribution")
plt.show()

# Age distribution by target
plt.figure(figsize=(8,4))
train_clean[train_clean["Heart Disease"]==0]["Age"].hist(alpha=0.6, bins=30)
train_clean[train_clean["Heart Disease"]==1]["Age"].hist(alpha=0.6, bins=30)
plt.title("Age Distribution (Absence vs Presence)")
plt.legend(["Absence", "Presence"])
plt.show()

from sklearn.model_selection import StratifiedKFold
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

X = train_clean.drop(columns=["Heart Disease"])
y = train_clean["Heart Disease"]

# Identify feature types
numeric_features = X.columns.tolist()

# For this dataset: all are numeric, but some behave like categorical.
# We'll keep them numeric for now (safe baseline).

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

preprocessor = ColumnTransformer(transformers=[
    ("num", numeric_transformer, numeric_features)
])

print("Total features:", len(numeric_features))
print("Features:", numeric_features)

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

# Model (strong clean baseline)
model = LogisticRegression(max_iter=2000)

pipe = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", model)
])

# CV
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

oof = np.zeros(len(X))
test_preds = np.zeros(len(test_clean))

for fold, (tr_idx, val_idx) in enumerate(skf.split(X, y), 1):
    X_tr, X_val = X.iloc[tr_idx], X.iloc[val_idx]
    y_tr, y_val = y.iloc[tr_idx], y.iloc[val_idx]

    pipe.fit(X_tr, y_tr)

    val_pred = pipe.predict_proba(X_val)[:, 1]
    oof[val_idx] = val_pred

    fold_auc = roc_auc_score(y_val, val_pred)
    print(f"Fold {fold} AUC: {fold_auc:.5f}")

    test_preds += pipe.predict_proba(test_clean)[:, 1] / skf.n_splits

cv_auc = roc_auc_score(y, oof)
print("\nOverall CV AUC:", cv_auc)


```

```python
submission = pd.DataFrame({
    "id": test_ids,
    "Heart Disease": test_preds
})

submission.to_csv("submission.csv", index=False)
submission.head()
```
