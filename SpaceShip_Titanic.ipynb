## üéØ What Is Our Main Task / Goal?

To **predict whether a passenger was transported to an alternate dimension or not (Yes / No).**

---

## üìÑ Column Details and Meanings

- **PassengerId**  
  A unique identifier for each passenger.

- **HomePlanet**  
  The planet from which the passenger comes.

- **CryoSleep**  
  Indicates whether the passenger was in suspended animation (sleeping) during the entire voyage.

- **Cabin**  
  Cabin information is stored in the format: `Deck / Number / Side`  
  - **Deck** ‚Üí Deck level  
  - **Number** ‚Üí Cabin number  
  - **Side** ‚Üí  
    - `P` = Port  
    - `S` = Starboard  

- **Destination**  
  The destination planet where the passenger is traveling.

- **Age**  
  Age of the passenger.

- **VIP**  
  Whether the passenger paid for special VIP services. in Yes or no

- **RoomService, FoodCourt, ShoppingMall, Spa, VRDeck**  
  Amount billed to the passenger for luxury amenities. Its an Money ok

- **Name**  
  Name of the passenger.

---

## üéØ Target Column

- **Transported**  
  Indicates whether the passenger was transported to another dimension.  
  - `True` ‚Üí Yes  
  - `False` ‚Üí No

```python
# Identifying Missing Values
# Action: Quantify the missing values in each column.

print("\n--- Missing Values Count per Column ---")
missing_values_count = df_train.isnull().sum()
display(missing_values_count[missing_values_count > 0].sort_values(ascending=False))

print("\n--- Percentage of Missing Values per Column ---")
missing_values_percent = (df_train.isnull().sum() / len(df_train)) * 100
display(missing_values_percent[missing_values_percent > 0].sort_values(ascending=False))

# Visualize missing data (useful for quick overall impression)
plt.figure(figsize=(12, 6))
sns.heatmap(df_train.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()

# What to look for:

# Columns with Missing Data: Which columns have NaN values?
# Extent of Missing Data: What percentage of values are missing?
# Small percentages (e.g., < 5-10%) might be easily imputed.
# Large percentages (e.g., > 50-70%) might suggest the column is not very useful or requires more advanced imputation/feature engineering.
# The heatmap provides a visual representation, where yellow/light colors typically indicate missing values, and dark colors indicate present data.
```

```python
# Unique Values and Categorical Features
# Action: Examine the number of unique values for each column, especially to identify categorical features and potential identifier columns.

print("\n--- Number of Unique Values per Column ---")
unique_counts = df_train.nunique().sort_values(ascending=False)
display(unique_counts)

print("\n--- Top 10 Columns with the Most Unique Values ---")
display(unique_counts.head(10))

print("\n--- Examples of Value Counts for Potential Categorical Columns (Top 5) ---")
# Let's pick a few columns that might be categorical based on `df.info()` or `nunique()`
# Adjust these based on your specific dataset's column names
potential_categorical_cols = df_train.select_dtypes(include='object').columns.tolist()
if len(potential_categorical_cols) == 0:
    print("No object type columns found.")
else:
    for col in potential_categorical_cols[:5]: # Display top 5 identified object columns
        print(f"\nValue counts for '{col}':")
        display(df_train[col].value_counts())
        if df_train[col].nunique() > 50: # If too many unique values, only show top 10
            print(f"(Only showing top 10 for '{col}' due to high cardinality)")
            display(df_train[col].value_counts().head(10))
# What to look for:

# High Unique Counts:
# Columns with unique == df_train.shape[0] are likely identifier columns (like id or customer_id), which usually aren't used directly for training but are important for linking data or submission.
# Columns with many unique values but not quite all unique might be high-cardinality categorical features (e.g., zip_code or product_name).
# Low Unique Counts: These are strong candidates for categorical features (e.g., gender, marital_status, product_category). We need to understand their distribution.
# Object Type Columns: Pay special attention to these. Are they truly categorical strings, or are they numbers that were incorrectly parsed due to special characters?
```

```python
# Target Variable Identification and Distribution
# This code examines the 'Transported' column to understand its type and distribution.

target_column = 'Transported'

# Check if the target column exists
if target_column not in df_train.columns:
    print(f"\nError: Target column '{target_column}' not found in the training data.")
else:
    print(f"\n--- Analysis of the Target Variable: '{target_column}' ---")

    # Check data type of the target variable
    print(f"Data type of '{target_column}': {df_train[target_column].dtype}")

    if df_train[target_column].dtype in ['int64', 'float64']:
        # It's a numerical target (regression problem)
        print("\nTarget variable is numerical (Regression Task).")
        print("Descriptive Statistics:")
        display(df_train[target_column].describe())

        # Visualize distribution
        plt.figure(figsize=(10, 6))
        sns.histplot(df_train[target_column].dropna(), kde=True)
        plt.title(f'Distribution of {target_column}')
        plt.xlabel(target_column)
        plt.ylabel('Frequency')
        plt.show()

        # Box plot for outliers
        plt.figure(figsize=(10, 6))
        sns.boxplot(y=df_train[target_column].dropna())
        plt.title(f'Box Plot of {target_column}')
        plt.ylabel(target_column)
        plt.show()

    else: # This block will execute for 'bool' and 'object' types that are categorical
        # It's a categorical target (classification problem)
        print("\nTarget variable is categorical (Classification Task).")
        print("Value Counts:")
        display(df_train[target_column].value_counts())
        print("\nValue Counts (Percentage):")
        display(df_train[target_column].value_counts(normalize=True) * 100)

        # Visualize distribution
        plt.figure(figsize=(10, 6))
        sns.countplot(data=df_train, x=target_column, order=df_train[target_column].value_counts().index)
        plt.title(f'Distribution of {target_column}')
        plt.xlabel(target_column)
        plt.ylabel('Count')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
```
